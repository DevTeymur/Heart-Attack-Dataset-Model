```python
def thresholds(df, col_name, q1=0.25, q3=0.75):
    quartile1 = df[col_name].quantile(q1)
    quartile3 = df[col_name].quantile(q3)
    IQR = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * IQR
    low_limit = quartile1 - 1.5 * IQR
    df.loc[(df[col_name] < low_limit), col_name] = int(np.round(df[col_name].mean()))
    df.loc[(df[col_name] > up_limit), col_name] = int(np.round(df[col_name].mean()))
    return df
```

```python
cols=['trtbps', 'chol', 'thalachh', 'oldpeak']
for col in cols:
    heart=thresholds(heart, col)
```

```python
for col in heart._get_numeric_data().columns:
    sns.boxplot(heart[col])
    plt.show()
```

# Building Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
```

```python
Y=heart.output
X=heart.drop(['output'], axis=1)
print(Y.shape, X.shape)
```

```python
models=[LogisticRegression(), RandomForestClassifier(), DecisionTreeClassifier(), SVC(), AdaBoostClassifier(), 
       GradientBoostingClassifier(), XGBClassifier(verbosity=0)] #47
```

```python
def best_randomstate(x,y, models):
    randomstates={}
    for model in models:
        scores={}
        for i in range(1,70):
            xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.25, random_state=i)
            model.fit(xtrain, ytrain)
            Y_test_pred=model.predict(xtest)
            scores[i]=np.round(accuracy_score(ytest, Y_test_pred), 4)
        randomstates[model]=max(scores, key=scores.get)
        print(model, 'done')
    return randomstates
```

```python
model_randomstate=best_randomstate(X, Y, models)
model_randomstate
```

```python
def training_models(x, y, models, randomstate):
    for model in models:
        X_train, X_test, Y_train, Y_test=train_test_split(x, y, test_size=0.25, random_state=randomstate.get(model))
        model.fit(X_train,Y_train)
        Y_test_pred=model.predict(X_test)
        print(model)
        print('Score: {}'.format(np.round(accuracy_score(Y_test, Y_test_pred), 4)))
        print('-----------------------------------------------------')
```

```python
training_models(X, Y, models, model_randomstate)
```

# Hyperparameter Tuning

```python
model_parameters={
    'Logistic Regression' : {
        'model': LogisticRegression(),
        'params': {'C': list(range(1,40)),
                   'penalty': ['l1', 'l2', 'elasticnet', 'none'],
                    'solver': ['saga', 'newton-cg', 'liblinear'],
                   'fit_intercept': [True, False]}},
    'Random Forest Classifier': {
        'model': RandomForestClassifier(),
        'params': {'n_estimators': list(range(10,50)),
                   'verbose': [0, 1],
                   'criterion': ['gini', 'entropy']}},
    'Decision Tree Classifier': {
        'model': DecisionTreeClassifier(),
        'params': {'criterion': ['gini','entropy'],
                  'splitter': ['best', 'random']}}, 
    'Ada Boost Classifier': {
        'model': AdaBoostClassifier(), 
        'params': {'n_estimators': list(range(40,60)), 
                   'algorithm': ['SAMME', 'SAMME.R']}},
    'Gradient Boosting Classifier': {
        'model': GradientBoostingClassifier(), 
        'params': {'learning_rate': [0.1, 0.2, 0.3, 0.01],
                   'n_estimators': list(range(95, 102)), 
                   'criterion': ['friedman_mse', 'mse', 'mae']}},
    'XGBoost Classifier': {
        'model': XGBClassifier(silent=True),
        'params': {'gamma': list(range(0, 3)), 
                   'max_depth': list(range(1, 10)),
                   'eta': [0,1, 0.3, 0.5, 0.7, 1]}},
}
```

```python
def results(model_name, Y_true, Y_pred, show=True):
    acc_score=np.round(accuracy_score(Y_pred, Y_true), 4)
    prec_score=np.round(precision_score(Y_pred, Y_true), 4)
    rec_score=np.round(recall_score(Y_pred, Y_true), 4)
    auc_roc_score=np.round(roc_auc_score(Y_pred, Y_true), 4)
    if show:
        print(model_name)
        print(f'Accuracy score: {acc_score}')
        print(f'Precision score: {prec_score}')
        print(f'Recall score: {rec_score}')
        print(f'Aoc Roc score: {auc_roc_score}')
        print('---------------------------------------')
    return acc_score, prec_score, rec_score, auc_roc_score
```

```python
def classification_models(x, y, model_params, model_randomstate):
    results_list=[]
    index=0
    for model_name, mp in model_params.items():
        X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=list(model_randomstate.values())[index])  
        clf=GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=True)
        clf.fit(X_train, Y_train)
        Y_test_pred=clf.predict(X_test)
        results(model_name, Y_test, Y_test_pred)
        a, b, c, d=results(model_name, Y_test, Y_test_pred, False)
        results_list.append({'model': model_name, 
                         'accuracy_score': a,
                         'precision_score': b,
                         'recall_score': c,
                         'auc_roc_score': d,
                         'best_score': clf.best_score_,
                         'best_params': clf.best_params_})
        index+=1
    return results_list
```

```python
result_list=classification_models(X, Y, model_parameters, model_randomstate)
```

```python
res=pd.DataFrame(result_list)
```

```python
res_sorted=res.sort_values('accuracy_score', ascending = False)
res_sorted
```

```python
res_sorted.to_csv('final_results.csv',index=False)
```
